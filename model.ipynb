```python
import tensorflow as tf
from tensorflow import keras
from keras import layers
from keras.models import Sequential
from keras.optimizers import Adam
```

```python
tf.config.list_physical_devices('GPU')
```

```python
train_data = tf.keras.utils.image_dataset_from_directory(
    '/content/drive/MyDrive/tiles', # update this
    validation_split = .2,
    subset='training',
    seed=0,
    image_size = (224, 224),
    batch_size=32,
)

val_data = tf.keras.utils.image_dataset_from_directory(
    '/content/drive/MyDrive/tiles',
    validation_split=.2,
    subset='validation',
    seed=0,
    image_size=(224,224),
    batch_size=32,
)
```

```python
resnet_mod = tf.keras.applications.ResNet50V2(
    weights=None,
    classes=1,
    classifier_activation="sigmoid",
)
```

```python
opt = keras.optimizers.Adam(use_ema=True)
```

```python
resnet_mod.compile(optimizer=opt, loss='binary_crossentropy', metrics=['accuracy'])
```

```python
history = resnet_mod.fit(
    x=train_data,
    validation_data = val_data,
    epochs = 25
)
```

```python
import matplotlib.pyplot as plt

acc = history.history['accuracy']
val_acc = history.history['val_accuracy']

loss = history.history['loss']
val_loss = history.history['val_loss']

epochs_range = range(50)

plt.figure(figsize=(8, 8))
plt.subplot(1, 2, 1)
plt.plot(epochs_range, acc, label='Training Accuracy')
plt.plot(epochs_range, val_acc, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.title('Training and Validation Accuracy')

plt.subplot(1, 2, 2)
plt.plot(epochs_range, loss, label='Training Loss')
plt.plot(epochs_range, val_loss, label='Validation Loss')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss')
plt.show()
```

```python
resnet_mod.save('airfields-resnet50v2-0.keras')
```
